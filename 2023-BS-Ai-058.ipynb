# Loops (all types)

# 1. for loop

```python
for i in range(1, 6):
    print(i)
```

# 2. While loop

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

# 3. Nested loop

```python
for i in range(1, 4):
    for j in range(1, 3):
        print(i, j)
```

# 4. Break and Continue

```python
for i in range(1, 6):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)
```

# Functions (types)

# 1. function with parameters and return

```python
def add(a, b):
    return a + b

print(add(5, 3))
```

# 2. lambda function

```python
square = lambda x: x * x
print(square(4))
```

# Lists (8–14 methods)

```python
my_list = [1, 2, 3, 4]

my_list.append(5)
my_list.extend([6, 7])
my_list.insert(1, 10)
my_list.remove(3)
my_list.pop()
my_list.sort()
my_list.reverse()
index = my_list.index(2)

print(my_list)
print("Index of 2:", index)
```

# Dictionaries:

# 1. List to Dictionary

```python
keys = ['name', 'age', 'city']
values = ['MAHAM', 20, 'Lahore']
my_dict = dict(zip(keys, values))
print(my_dict)
```

# 2. Dictionary to List

```python
dict_items = list(my_dict.items())
print(dict_items)
```

# 3. Tuple

```python
my_tuple = (1, 2, 3)
print(my_tuple)
```

# Heading 3: Matrix Multiplication

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

result = np.dot(A, B)
print(result)
```
